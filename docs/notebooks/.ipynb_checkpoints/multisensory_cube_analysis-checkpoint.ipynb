```python
from multisensory_cube_analysis import test_burst_amplification#, plot_ampli_on_dist

def plot_ampli_on_dist(datafile, max_ampl=10):
    df=pd.read_csv(datafile)
    fig=plt.figure(figsize=(15, 6))

    ax=plt.subplot(1, 1, 1)

    legend=list()

    for i in range(max_ampl+1):
        ax.plot(df["dist"], df[f"{i}-th amplification"], '-', alpha=1)
        legend.append(f"{i}-th amplification")

    ax.legend(legend)

    target="[{}, {}, {}]".format(
        df["target_1"][0], df["target_2"][0], df["target_3"][0])
    ax.set_title(
        f"Average bursts - Target = {target} (dataset = \"{datafile}\")")
    ax.set_xlabel(f"Euclidean distance from {target}")
    ax.set_ylabel("Average burst")


print("\n")
test_burst_amplification(BurstAModel(3,1), max_ampl=3, target=[0, 0, 0], sample_ratio=5, shots = pow(10, 2))
plot_ampli_on_dist("data/df_BurstAmplification_n3_tau1_target[0, 0, 0]_sr5.csv", max_ampl=3)
plt.show()
plot_all_amplifications_4d("data/df_BurstAmplification_n3_tau1_target[0, 0, 0]_sr5.csv", max_ampl=3)
plt.show()


print("\n")
test_burst_amplification(BurstAModel(3,1), max_ampl=3, target=[.5,.5,.5], sample_ratio=5, shots = pow(10, 2))
plot_ampli_on_dist(df_burst_center, max_ampl=10)
plt.show()
```
