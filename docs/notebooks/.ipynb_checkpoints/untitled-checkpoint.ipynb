```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from qrobot.models import AngularModel, LinearModel
```

```python
def input_dataset(sample_ratio=1, n=3):
    
    from itertools import combinations_with_replacement, permutations
    
    # Generate the range of values
    values = np.arange(0, 1, 1/sample_ratio)
    # Add the other range bound
    values = np.append(values,1)
    
    # Avoid results like 0.30000000000000004
    values = [round(x,10) for x in values]

    # Then generate all the possible combinations of n of these values
    # and their permutations
    data_list = list()
    comb = combinations_with_replacement(values, n)
    for c in comb:
        perm = permutations(c)
        for p in perm:
            data_list.append(p)

    # Delete then the duplicate elements
    dataset = list(dict.fromkeys(data_list))
    
    # Convert it as a dataframe
    df = pd.DataFrame(dataset, columns=['x1', 'x2', 'x3'])
    
    #df.sort_values(by=['x1'], inplace=True)
    
    return df
```

```python
input_dataset(1)
```

```python
X = [x for x in range(1,50)]
Y = [len(input_dataset(x)) for x in X]
plt.figure(figsize=(15,7), dpi=150)
plt.xlabel("Sample ratio")
plt.ylabel("Dataset size")
plt.plot(X,Y)
plt.legend(["Dataset size"])
plt.grid()
plt.show()
```

```python
model_ang = AngularModel(3,1)
model_lin = LinearModel(3,1)
```

For the progress bar ([reference](https://towardsdatascience.com/progress-bars-in-python-4b44e8a4c482)):

```python
# pip install ipywidgets 
# jupyter nbextension enable --py widgetsnbextension
# jupyter labextension install @jupyter-widgets/jupyterlab-manager
from tqdm.notebook import tqdm
```

```python
def test_model(model, sample_ratio = 10, target=False):
    
    df = input_dataset(sample_ratio)
    print("Dataset generated!")

    states_list = ["000", "001", "010", "100", "011", "110", "101", "111"]
    shots = pow(10, 5)
    result = list()

    iteration = 1
    tot_iter = len(df)
    print(f"The dataset contains {tot_iter} entries.")

    with tqdm(total=tot_iter) as pbar:
        for index, row in df.iterrows():
            
            # Start by clearing the model
            model.clear()
            
            # Then encode the inpiut
            for i in range(model.n):
                model.encode(row[i],i+1)  # dimensions start at 1, not at 0!
                
            if target:
                model.query(target)
            else:
                # Use the origin as target (for dist computation)
                target = [0,0,0]

            # Get the distance from the target for this input [x1, x2, x3]
            dist = np.linalg.norm(np.array(row)-np.array(target))
            # The distance is normalized on the hypercube's diagonal
            dist = dist/np.sqrt(model.n)

            # Simulate the measurement counts
            counts = model.measure(shots)

            # Convert counts in probabilities
            probabilities = list()
            for key in states_list:
                try:
                    probabilities.append(counts[key]/shots)
                except KeyError:
                    probabilities.append(0)

            # Store probabilities and distance in the result list
            probabilities.append(dist)
            result.append(probabilities)
            
            iteration += 1
            pbar.update(1) # Update progress bar

    # Add labels
    result_cols = states_list
    result_cols.append(f"Dist from {target}")
    # Convert result list to a dataframe
    df_result = pd.DataFrame(result, columns=result_cols)

    # Attach the result to the input dataframe
    df = pd.concat([df, df_result], axis=1)
    
    return df 
```

```python
sample_ratio = 15
test_model(model_ang, sample_ratio).to_csv(
    "data/df_angular_step"+str(sample_ratio)+".csv", index=False)

target=[.5,.5,.5]
test_model(model_ang, sample_ratio, target).to_csv(
    "data/df_angular_step"+str(sample_ratio)+"_target.csv", index=False)
```

```python
sample_ratio = 15
test_model(model_lin, sample_ratio).to_csv(
    "data/df_linear_step"+str(sample_ratio)+".csv", index=False)

target=[.5,.5,.5]
test_model(model_lin, sample_ratio, target).to_csv(
    "data/df_linear_step"+str(sample_ratio)+"_target.csv", index=False)
```

```python
fig = plt.figure()

def data_4dplot(datafile, fig=plt.figure()):
    df = pd.read_csv(datafile)
    ax = fig.add_subplot(111, projection='3d')

    ax.set_title(f"{datafile}")

    # 3 Axis (3 dimensions)
    x = df["x1"]
    y = df["x2"]
    z = df["x3"]
    # Colour (4th dimension)
    c = df.iloc[:,11]

    img = ax.scatter(x, y, z, c=c, cmap=plt.get_cmap('RdBu'),  vmin=0, vmax=1)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_zlabel('x3')
    ax.set_xlim([0,1])
    ax.set_ylim([0,1])
    ax.set_zlim([0,1])

    # target_rgb = mapper.base[mapper.state2idx[target_state]-1]
    # ax.scatter(target_rgb[0], target_rgb[1], target_rgb[2], c="blue", linewidths=20, marker="P")

    fig.colorbar(img)
    plt.show()
```
